```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
bike_buyers = pd.read_csv('../Data/bike_buyers.csv')
```

```python
bike_buyers.head()
```

```python
#remove NA values
bike_buyers = bike_buyers.dropna()
```

```python
bike_buyers.groupby('Income').count()['Purchased Bike'].plot(kind='bar', title='Number of Bikes Purchased by Income',
                                                             ylabel="Number of Bikes Purchased", xlabel="Income",color="#549c4c")
plt.show()
```

```python
#show a bar graph of marital status with purchased bikes
bike_buyers.groupby('Marital Status').count()['Purchased Bike'].plot(kind='bar', title='Number of Bikes Purchased by Marital Status',
                                                                ylabel="Number of Bikes Purchased", xlabel="Marital Status",color="#549c4c")
plt.show()
```

```python
#show a graph describing how much each education win as income
bike_buyers.groupby('Education').mean()['Income'].plot(kind='bar', title='Average Income by Education',
                                                                ylabel="Average Income", xlabel="Education",color="#549c4c")
plt.show()
```

```python
#show a graph comparing the amount of Cars of each occupation
bike_buyers.groupby('Occupation').count()['Cars'].plot(kind='bar', title='Number of Cars by Occupation',
                                                                ylabel="Number of Cars", xlabel="Occupation",color="#549c4c")
plt.show()
```
