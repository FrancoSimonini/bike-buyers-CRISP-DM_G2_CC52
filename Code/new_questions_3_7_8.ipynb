```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
bike_buyers = pd.read_csv('../Data/bike_buyers.csv')
```

```python
bike_buyers.head()
```

```python
#remove NA values
bike_buyers = bike_buyers.dropna()
```

```python
bike_buyers.groupby('Income').count()['Purchased Bike'].plot(kind='bar', title='Number of Bikes Purchased by Income',
                                                             ylabel="Number of Bikes Purchased", xlabel="Income",color="#549c4c")
plt.show()
```

```python
#show a bar graph of marital status with purchased bikes
bike_buyers.groupby('Marital Status').count()['Purchased Bike'].plot(kind='bar', title='Number of Bikes Purchased by Marital Status',
                                                                ylabel="Number of Bikes Purchased", xlabel="Marital Status",color="#549c4c")
plt.show()
```

```python
#categorize income into low, medium, and high
#bike_buyers['Income'] = pd.cut(bike_buyers['Income'], bins=[0, 30000, 75000, 150000], labels=['Low', 'Medium', 'High'])
#show a graph of income vs education
bike_buyers.groupby('Income').count()['Education'].plot(kind='bar', title='Number of Bikes Purchased by Income',
                                                                ylabel="Number of Bikes Purchased", xlabel="Income",color="#549c4c")
plt.show()
```

```python
#show a grapgh of numbers of cars per Occupation
bike_buyers.groupby('Occupation').count()['Cars'].plot(kind='bar', title='Number of Bikes Purchased by Occupation',
                                                                ylabel="Number of Bikes Purchased", xlabel="Occupation",color="#549c4c")
plt.show()
```
